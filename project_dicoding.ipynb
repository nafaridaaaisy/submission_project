### Cleaning Data

```python
# Cleaning data geolocation_df

# removing duplicates
geolocation_df.drop_duplicates(inplace=True)

# checking wether there are still duplicates or not
print("Jumlah duplikasi: ", customers_df.duplicated().sum())
```

```python
# Cleaning data order_reviews_df

# filtering missing values
order_reviews_df[order_reviews_df.review_comment_title.isna()]
order_reviews_df[order_reviews_df.review_comment_message.isna()]

# replace the missing values (NaN) with no comments text
order_reviews_df.fillna(value="(no comments)", inplace=True)

# checking
order_reviews_df.isna().sum()
```

```python
# Cleaning data order_reviews_df

# changing the datatype of review_creation_date and review_answer_timestamp to be datetime type
datetime_columns = ["review_creation_date", "review_answer_timestamp"]

for column in datetime_columns:
  order_reviews_df[column] = pd.to_datetime(order_reviews_df[column])
# checking the info of this data
order_reviews_df.info()
```

```python
# Cleaning data orders_df

# changing the datatype of review_creation_date and review_answer_timestamp to be datetime type
orders_datetime_columns = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]

for column in orders_datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])

# checking the info of this data
orders_df.info()
```

```python
# Cleaning data orders_df

# filtering missing values
orders_df[orders_df.order_approved_at.isna()]
orders_df[orders_df.order_delivered_carrier_date.isna()]
orders_df[orders_df.order_delivered_customer_date.isna()]
```

**Insight:**
- Our business question is designed to ask especially about the sales performance and revenue
- So we decide to use only the order datas that have been approved by the seller

```python
# We drop the missing value (NaT) from the section of order_approved_at
orders_df = orders_df.dropna(subset=['order_approved_at'])

# checking wether there are still missing values or not
orders_df.isna().sum()
```

**Insight:**
- From the CSV file display of orders_df, we know that the time for carrier_date is variative
- Let's assume that the order_carrier_date is conducted in the sameday of the order approval

```python
# replace the missing value of delivered_carrier_date with the same datetime of order_approved_at
orders_df['order_delivered_carrier_date'] = orders_df['order_delivered_carrier_date'].fillna(orders_df['order_approved_at'])

# checking wether there are still missing values or not
orders_df.isna().sum()
```

**Insight:**
- From the CSV file display of orders_df, we know that the time for order_delivered_customer_date is variative
- Let's assume the delivered_customer_date is decided maximum on the same date of order_estimated_delivery_date

```python
# replace the missing value of order_delivered_customer_date with the same datetime of order_estimated_delivery_date
orders_df['order_delivered_customer_date'] = orders_df['order_delivered_customer_date'].fillna(orders_df['order_estimated_delivery_date'])

# checking wether there are still missing values or not
orders_df.isna().sum()
```

```python
# Cleaning data products_df

# filtering missing values of product dimension
products_df[products_df.product_weight_g.isna()]
products_df[products_df.product_length_cm.isna()]
products_df[products_df.product_height_cm.isna()]
products_df[products_df.product_width_cm.isna()]
```

**Insight:**
- Actually, we can't precisely decide the dimension (weight, length, height, and width) of the product only based on statistical data stuffs.
- It is because the product type is unique
- We can can impute missing values by randomly selecting values from the distribution of other available product dimensions.
- This preserves the variation in the data.

```python
# fill the missing value (NaN) from the section of product_category_name with random sampling techniques
columns_to_fill = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
products_df.loc[products_df[columns_to_fill].isnull().any(axis=1), columns_to_fill] = products_df[columns_to_fill].dropna().sample(1).values

# checking wether there are still missing values or not
products_df.isna().sum()
```

```python
# Cleaning data products_df

# filtering missing values of product_category_name, product_name_lenght, product_description_lenght, and product_photos_qty
products_df[products_df.product_category_name.isna()]
products_df[products_df.product_name_lenght.isna()]
products_df[products_df.product_description_lenght.isna()]
products_df[products_df.product_photos_qty.isna()]
```

**Insight:**
- We replace the product_category_name with "uncategorized_product"
- We can't drop it because the product has their own identity (product_id)
- The product_name_length should be filled by the length of the product_category_name
- The product_description_length should be replaced by the length of 0 (it means there's no description for the product)
- The products photos qty can be filled by the mode number of the photo, because teh QTY should be discrete number. 50% of the data still show that the quantity of the number is 1, the 75% show 3, the std is around 1.7, mean is around 2, while the max qty is 20. So to maintain the accuracy and the variety of the data, we can choose to fill it with mode.

```python
# fill the missing value (NaN) from the section of product_category_name with "uncategorized product"
products_df['product_category_name'] = products_df['product_category_name'].fillna(value="uncategorized product")

# fill the missing value (NaN) from the section of product_name_lenght with the length of "undescribed product"
replacement_string = "undescribed product"            # define the string "undescribed product"
replacement_length = len(replacement_string)          # calculate the length of the string "undescribed product"
products_df['product_name_lenght'] = products_df['product_name_lenght'].fillna(replacement_length)  # fill NaN in the product_name_length column with the length of the string "undescribed product"

# fill the missing value (NaN) from the section of product_description_lenght with the length of "no description"
replacement_string1 = "no description"                 # define the string "no description"
replacement_length1 = len(replacement_string1)         # calculate the length of the string "undescribed product"
products_df['product_description_lenght'] = products_df['product_description_lenght'].fillna(replacement_length1)  # fill NaN in the product_name_length column with the length of the string "undescribed product"

# fill the missing value (NaN) from the section of product_photos_qty with mode
mode_value = products_df['product_photos_qty'].mode()[0]           #calculate the mode, .mode() returns a series, so we take the first value
products_df['product_photos_qty'].fillna(mode_value, inplace=True) #fill missing values (NaN) with the mode

# checking the missing value info of this data
products_df.isna().sum()
```
